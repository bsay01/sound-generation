```python
import numpy as np
import soundfile as sf
import IPython.display as ipd
from sg_functions import *
```

```python
tone_freq = [185]
tone_amp = 1
tone_duration = 0.1

tone_ADSR = generate_ADSR_envelope(tone_duration, 0.00, 0.127, 0.910, 0.2)
tone_ADSR_len = len(tone_ADSR)
tone_ADSR_duration = tone_ADSR_len / SAMPLE_RATE

tone_sin = 0*tone_ADSR
for f_i in tone_freq:
    tone_sin = np.add(tone_sin, generate_sine(note_num=freq2midi(f_i), duration=tone_ADSR_duration, sr=SAMPLE_RATE))

tone_r_ADSR = np.array([i/100 for i in tone_ADSR])
tone_sin = np.multiply(tone_sin, tone_r_ADSR)[0:tone_ADSR_len]

ipd.Audio(tone_sin, rate=SAMPLE_RATE)
```

```python
noise_duration_ratio = 0.9

noise_ADSR = generate_ADSR_envelope(tone_duration * noise_duration_ratio, 0.00, 0.1, 0.15, 0.15)
noise_ADSR_len = len(noise_ADSR)
noise_ADSR_duration = noise_ADSR_len / SAMPLE_RATE

noise = np.random.uniform(-1, 1, int(SAMPLE_RATE * noise_ADSR_duration))

noise_r_ADSR = np.array([i/100 for i in noise_ADSR])
noise = np.multiply(noise, noise_r_ADSR)[0:noise_ADSR_len]

# add silence to the end to match the length of the snare tone
noise = np.append(noise, np.array([0 for i in tone_sin]))[0:len(tone_sin)]

ipd.Audio(noise, rate=SAMPLE_RATE)
```

```python
snare = np.add(0.6*tone_sin, noise)

ipd.Audio(snare, rate=SAMPLE_RATE)
```

```python
# Save to .wav file
try:
    sf.write("data_drums/" + "snare_drum.wav", snare, SAMPLE_RATE)
except:
    sf.write("snare_drum.wav", snare, SAMPLE_RATE)
```
